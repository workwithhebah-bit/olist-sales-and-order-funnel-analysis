```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine=create_engine('mysql+pymysql://root:Opensql@localhost/olist_db')
print('Connected to MySQL!')
```

```python
path='/Users/hebah/Olist_Project/Data/'

orders = pd.read_csv(path + 'olist_orders_dataset.csv')
customers = pd.read_csv(path + 'olist_customers_dataset.csv')
items = pd.read_csv(path + 'olist_order_items_dataset.csv')
payments = pd.read_csv(path + 'olist_order_payments_dataset.csv')
reviews = pd.read_csv(path + 'olist_order_reviews_dataset.csv')
products = pd.read_csv(path + 'olist_products_dataset.csv')
sellers = pd.read_csv(path + 'olist_sellers_dataset.csv')
translation = pd.read_csv(path + 'product_category_name_translation.csv')

print("All files loaded!")
print("Orders shape:", orders.shape)
```

```python
orders.to_sql('orders', engine, if_exists='replace', index=False)
print("orders done")

customers.to_sql('customers', engine, if_exists='replace', index=False)
print("customers done")

items.to_sql('items', engine, if_exists='replace', index=False)
print("items done")

payments.to_sql('payments', engine, if_exists='replace', index=False)
print("payments done")

reviews.to_sql('reviews', engine, if_exists='replace', index=False)
print("reviews done")

products.to_sql('products', engine, if_exists='replace', index=False)
print("products done")

sellers.to_sql('sellers', engine, if_exists='replace', index=False)
print("sellers done")

translation.to_sql('translation', engine, if_exists='replace', index=False)
print("translation done")

print("\n All tables loaded into MySQL!")
```

```python
import pandas as pd
from sqlalchemy import create_engine


engine = create_engine('mysql+pymysql://root:Opensql@127.0.0.1:3306/olist_db')


sql = (
    "SELECT o.order_id, o.customer_id, o.order_status, "
    "o.order_purchase_timestamp, o.order_approved_at, "
    "o.order_delivered_customer_date, o.order_estimated_delivery_date, "
    "c.customer_city, c.customer_state, "
    "pay.payment_type, pay.payment_value, "
    "r.review_score, i.price, i.freight_value, "
    "t.product_category_name_english "
    "FROM orders o "
    "LEFT JOIN customers c ON o.customer_id = c.customer_id "
    "LEFT JOIN payments pay ON o.order_id = pay.order_id "
    "LEFT JOIN reviews r ON o.order_id = r.order_id "
    "LEFT JOIN items i ON o.order_id = i.order_id "
    "LEFT JOIN products p ON i.product_id = p.product_id "
    "LEFT JOIN translation t ON p.product_category_name = t.product_category_name "
    "WHERE o.order_status != 'canceled' "
    "AND pay.payment_value IS NOT NULL"
)

# Run query
df = pd.read_sql(sql, engine)
print("Done!")
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
df.head()
```
